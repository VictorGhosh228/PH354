```python
import numpy as np
M=0.0
L=3
for i in range(-L,L+1):
  for j in range(-L,L+1):
    for k in range(-L,L+1):
      if i==0 and j==0 and k==0:
        continue
      else:
        b=(i+j+k)
        M+=(-1)**b/(np.sqrt(i**2 +j**2 +k**2))
        print(M)
print(M)
```
